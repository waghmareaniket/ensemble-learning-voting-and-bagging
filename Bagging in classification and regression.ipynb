```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
x,y=make_moons(n_samples=1000,noise=0.2)
```

```python
df={"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
data=pd.DataFrame(df)
```

```python
sns.scatterplot(x="x1",y="x2",hue="y",data=data)
plt.show()
```

```python
x_a=data.iloc[:,:-1]
```

```python
y_a=data["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_a,y_a,random_state=42,test_size=0.2)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC(),n_estimators=30)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train)*100
```

```python
bg.score(x_test,y_test)*100
```

```python
rf=RandomForestClassifier(n_estimators=30)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)*100, rf.score(x_test,y_test)*100
```
